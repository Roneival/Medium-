```python
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic ECG signal
duration = 3  # Reduced duration to 5 seconds
sampling_rate = 1000  # Sampling rate in Hz
ecg_signal = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate, noise=0.01)

# Simulate power line interference noise
pli_freq = 50  # Power line frequency in Hz
pli_noise = 0.05 * np.sin(2 * np.pi * pli_freq * np.linspace(0, duration, duration * sampling_rate))

# Combine ECG signal with power line interference
combined_signal = ecg_signal + pli_noise

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot the clean synthetic ECG signal
axs[0].plot(ecg_signal, color='blue')
axs[0].set_title('Synthetic ECG Signal')
axs[0].set_xlabel('Time (samples)')
axs[0].set_ylabel('Amplitude')
axs[0].grid()

# Plot the ECG signal with Power Line Interference
axs[1].plot(combined_signal, color='orange')
axs[1].set_title('ECG with Power Line Interference')
axs[1].set_xlabel('Time (samples)')
axs[1].set_ylabel('Amplitude')
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic ECG signal
duration = 2  # Duration in seconds
sampling_rate = 1000  # Sampling rate in Hz
ecg_signal = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate, noise=0.01)

# Simulate motion artifact
np.random.seed(0)  # Seed for reproducibility
motion_artifact = np.random.normal(0, 0.1, size=ecg_signal.shape)  # Random noise simulating motion artifact

# Combine ECG signal with motion artifact
motion_signal = ecg_signal + motion_artifact

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot the clean synthetic ECG signal
axs[0].plot(ecg_signal, color='blue')
axs[0].set_title('Clean ECG Signal')
axs[0].set_xlabel('Time (samples)')
axs[0].set_ylabel('Amplitude')
axs[0].grid()

# Plot the ECG signal with Motion Artifact
axs[1].plot(motion_signal, color='coral')
axs[1].set_title('ECG Signal with Motion Artifact')
axs[1].set_xlabel('Time (samples)')
axs[1].set_ylabel('Amplitude')
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic ECG signal
duration = 5  # Duration in seconds
sampling_rate = 1000  # Sampling rate in Hz
ecg_signal = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate, noise=0.01)

# Simulate muscle contraction (EMG) noise
peak_to_peak_amplitude = np.ptp(ecg_signal)
std_dev_emg_noise = 0.1 * peak_to_peak_amplitude  # 10% of the peak-to-peak amplitude
duration_emg_noise = int(0.05 * sampling_rate)  # 50 ms of noise

# Generate EMG noise at random intervals
emg_noise = np.zeros_like(ecg_signal)
for _ in range(duration * 20):  # Assuming an average of 20 contractions per second
    start = np.random.randint(0, len(ecg_signal) - duration_emg_noise)
    emg_noise[start:start + duration_emg_noise] += np.random.normal(0, std_dev_emg_noise, duration_emg_noise)

# Combine ECG signal with EMG noise
combined_signal = ecg_signal + emg_noise

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 8))

# Plot the clean synthetic ECG signal
axs[0].plot(ecg_signal, color='blue')
axs[0].set_title('Clean ECG Signal')
axs[0].set_xlabel('Time (samples)')
axs[0].set_ylabel('Amplitude')
axs[0].grid()

# Plot the ECG signal with Muscle Contraction (EMG) Noise
axs[1].plot(combined_signal, color='skyblue')
axs[1].set_title('ECG Signal with Muscle Contraction (EMG) Noise')
axs[1].set_xlabel('Time (samples)')
axs[1].set_ylabel('Amplitude')
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
import neurokit2 as nk
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic ECG signal
ecg_signal = nk.ecg_simulate(duration=1, sampling_rate=1000)  # 5-second long ECG signal

# Simulate Baseline Wander
baseline_wander = np.cumsum(np.random.normal(0, 0.15 * np.ptp(ecg_signal), size=ecg_signal.shape))

# Add Baseline Wander to the ECG signal
ecg_with_baseline_wander = ecg_signal + baseline_wander

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(ecg_signal, label='Original ECG Signal')
plt.title('Original ECG Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(ecg_with_baseline_wander, label='ECG with Baseline Wander', color='orange')
plt.title('ECG Signal with Baseline Wander')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import neurokit2 as nk

# Parameters
duration = 5  # Duration in seconds
sampling_rate = 1000  # Sampling rate in Hz

# Generate a clean ECG signal
ecg_clean = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate, heart_rate=60)

# Create Baseline Wander noise
baseline_wander = nk.signal_simulate(duration=duration, frequency=0.5, amplitude=0.3 * np.ptp(ecg_clean), sampling_rate=sampling_rate)

# Create 50 Hz Power Line interference noise
powerline_noise = np.sin(2 * np.pi * 50 * np.linspace(0, duration, duration * sampling_rate))

# Create EMG noise
emg_noise = np.random.normal(0, 0.01, size=len(ecg_clean)) * np.ptp(ecg_clean)

# Combine the signals with noise
ecg_baseline_wander = ecg_clean + baseline_wander
ecg_powerline_noise = ecg_clean + powerline_noise
ecg_emg_noise = ecg_clean + emg_noise

# Plotting
plt.figure(figsize=(15, 9))

# Clean ECG
plt.subplot(3, 1, 1)
plt.plot(ecg_clean, label='Clean ECG', color='blue')
plt.title('Clean ECG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# ECG with Baseline Wander
plt.subplot(3, 1, 2)
plt.plot(ecg_baseline_wander, label='ECG with Baseline Wander', color='blue')
plt.title('ECG Signal with Baseline Wander')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# ECG with Power Line Interference
plt.subplot(3, 1, 3)
plt.plot(ecg_powerline_noise, label='ECG with 50 Hz Power Line Noise', color='blue')
plt.title('ECG Signal with 50 Hz Power Line Interference')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
